## Bollinger Bands



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import quandl
```

```python
sbi=quandl.get('NSE/SBIN',start_date='2017-01-01',end_date='2018-01-01')
sbi.tail()
```

```python
sbi['Close:20 day Mean']=sbi['Close'].rolling(window=20).mean()
sbi['Upper']=sbi['Close:20 day Mean']+2*sbi['Close'].rolling(window=20).std()
sbi['Lower'] = sbi['Close:20 day Mean'] - 2*sbi['Close'].rolling(window=20).std()
sbi[['Close','Close:20 day Mean','Upper','Lower']].plot(figsize=(16,6))
```
